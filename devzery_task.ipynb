### Tasks to perform

```python
# Create tasks for your agents
task1 = Task(
  description="""Give a complete report showing time complexity, space complexity and different optimised coding approach for the given backstory. write down code""",
  agent=coder
)
```

```python
task2 = Task(
  description="""Using the insights provided, develop a code showing different ways it can be made better""",
  agent=writer
)
```

### Create a Crew

```python

# Instantiate your crew with a sequential process
crew = Crew(
  agents=[coder, writer],
  tasks=[task1, task2],
  verbose=2, # You can set it to 1 or 2 to different logging levels
)
```

### Kickoff the crew - let the magic happen

```python
# Get your crew to work!
result = crew.kickoff()
```
